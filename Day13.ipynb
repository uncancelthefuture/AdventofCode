```python
import numpy as np

f = open("13-input.txt", "r")
data = f.read().replace('.','0').replace('#', '1').strip().split("\n\n")
f.close()
d = {'.':0,'#':1}

data = [np.array([list(y) for y in x.split("\n")], dtype=int) for x in data]
```

```python
def iterate_vert_symmetry(map, i):
  if i < 0 or i+1 == len(map[:,0]):
    return True
  elif np.array_equal(map[i,:], map[i+1,:]):
    map = np.delete(map,[i,i+1], axis=0)
    return iterate_vert_symmetry(map, i-1)
  return False

def iterate_horz_symmetry(map,i):
  if i < 0 or i+1 == len(map[0,:]):
    return True
  elif np.array_equal(map[:,i], map[:,i+1]):
    map = np.delete(map,[i,i+1], axis=1)
    return iterate_horz_symmetry(map, i-1)
  return False

def horz_symm(map):
  m = 0
  for i in range(len(map[0,:]) - 1):
    if np.array_equal(map[:,i], map[:,i+1]):
      if iterate_horz_symmetry(map,i):
        m = i + 1
        return m
  return 0

def vert_symm(map):
  m = 0
  for i in range(len(map[:,0]) - 1):
    if np.array_equal(map[i,:], map[i+1,:]):
      if iterate_vert_symmetry(map,i):
        m = i + 1
        return m*100
  return 0


s = 0
for map in data:
  s += horz_symm(map) + vert_symm(map)
s

```

```python
def vert_mismatch(map,i):
  m1 = map[i,:]
  m2 = map[i+1,:]
  m = (m1 + m2) % 2
  return m.sum()

def iterate_vert_symmetry(map, i, smudge):
  if (i < 0 or i+1 == len(map[:,0])):
    if smudge:
      return True
    else:
      return False
  elif vert_mismatch(map, i) == 1 and not smudge:
    smudge = 1
    map = np.delete(map,[i,i+1], axis=0)
    return iterate_vert_symmetry(map, i-1, smudge)
  elif vert_mismatch(map,i) == 0:
    map = np.delete(map,[i,i+1], axis=0)
    return iterate_vert_symmetry(map, i-1, smudge)
  return False

def vert_symm(map):
  m = 0
  for i in range(len(map[:,0]) - 1):
    if vert_mismatch(map, i) < 2:
      if iterate_vert_symmetry(map,i,0):
        m = i + 1
        return m*100
  return 0

def horz_mismatch(map,i):
  m1 = map[:,i]
  m2 = map[:,i+1]
  m = (m1 + m2) % 2
  return m.sum()

def iterate_horz_symmetry(map, i, smudge):
  if (i < 0 or i+1 == len(map[0,:])):
    if smudge:
      return True
    else:
      return False
  elif horz_mismatch(map, i) == 1 and not smudge:
    smudge = 1
    map = np.delete(map,[i,i+1], axis=1)
    return iterate_horz_symmetry(map, i-1, smudge)
  elif horz_mismatch(map,i) == 0:
    map = np.delete(map,[i,i+1], axis=1)
    return iterate_horz_symmetry(map, i-1, smudge)
  return False

def horz_symm(map):
  m = 0
  for i in range(len(map[0,:]) - 1):
    if horz_mismatch(map, i) < 2:
      if iterate_horz_symmetry(map,i,0):
        m = i + 1
        return m
  return 0

s = 0

for map in data:
  s += vert_symm(map) + horz_symm(map)

s

```
