```python
f = open("11-input.txt", "r")

data = f.read().strip().split("\n")

f.close()

data

width = len(data[0])
height = len(data)
galaxies = set()

for i in range(width):
  for j in range(height):
    if data[j][i] == '#':
      galaxies.add((i,j))

cols = [x for x in range(width)]
rows = [x for x in range(height)]

for x in galaxies:
  try:
    rows.remove(x[1])
  except:
    ValueError
  try:
    cols.remove(x[0])
  except:
    ValueError



galaxies1 = set([])
for x in galaxies:
  r = 0
  h = 0
  for row in rows:
    if row < x[1]:
      r += 1
    else:
      break
  for col in cols:
    if col < x[0]:
      h += 1
    else:
      break
  x1 = (x[0] + h, x[1] + r)
  galaxies1.add(x1)

galaxies1

def dist(x,y):
  return abs(x[0]-y[0]) + abs(x[1]-y[1])
s = 0
for x in galaxies1:
  for y in galaxies1:
    s += dist(x,y)

s = s//2

s

```
