```python
import re

f = open("/content/10-input", "r")

data = f.read().strip().split("\n")

f.close()

N = len(data[0])

def pipe(x):
  u = (0,-1)
  d = (0,1)
  l = (-1,0)
  r = (1,0)
  match x:
    case '|':
      return [u,d]
    case '-':
      return [r, l]
    case 'L':
      return [u, r]
    case 'J':
      return [u,l]
    case '7':
      return [l,d]
    case 'F':
      return [r,d]
    case 'S':
      return [u,r,d,l]
    case '.':
      return []

def v_add(x,y):
  return (x[0]+y[0], x[1]+y[1])

def next_pipe(x,d, last):
  pipes = []
  dir = pipe(d[x])
  for v in dir:
    x1 = v_add(x,v)
    if x1 == last:
      pass
    else:
      dir1 = pipe(d[x1])
      for v1 in dir1:
        if v_add(v1,v) == (0,0):
          pipes.append(x1)
  return pipes

d = {}
loop = []
for j in range(N):
  for i in range(len(data)):
    if data[i][j] == 'S':
      pos = (j,i)
    d[(j,i)] = data[i][j]

last = None
s = 0
start = 1
x = pos
loop = [x]
while d[x] != 'S' or start:
  start = 0
  s += 1
  x1 = x
  x = next_pipe(x,d,last)[0]
  last = x1
  loop.append(x)


s = s//2
s
```

```python
last = loop[0]
vertex_set = []
i = 1
for x in loop[1:]:
  if d[x] not in {'-','|'}:
    if d[x] == 'S':
      l = next_pipe(x,d,None)
      if v_add(l[0],l[1]) == (0,0):
        pass
      else:
        vertex_set.append(x)
    else:
      vertex_set.append(x)

x = vertex_set[-1]
y = vertex_set[0]
area = (x[1]+y[1])*(x[0]-y[0])
for i in range(len(vertex_set)-1):
  x = vertex_set[i]
  y = vertex_set[i+1]
  area += (x[1]+y[1])*(x[0]-y[0])
area = abs(area // 2)

i = area - s + 1

i
```
