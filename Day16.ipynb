```python
import numpy as np
import sys
from collections import defaultdict
sys.setrecursionlimit(10000)
```

```python
f = open("16-input.txt", "r")

data = [list(x) for x in f.read().strip().split('\n')]

f.close()

data  = np.array(data)

map ={}
N = len(data)
M = len(data[0])
for i in range(N):
  for j in range(M):
    map[(i,j)] = data[i][j]

class DefaultDict(dict):
  __missing__ = lambda self, key: key[0]
```

```python
dir = {'>':(0,1), '<':(0,-1), '^':(-1,0), 'v':(1,0)}
mirrors = {('>','|')  : ['^','v'],
           ('<','|')  : ['^','v'],
           ('^','-')  : ['>','<'],
           ('v','-')  : ['>','<'],
           ('>', '\\'): ['v'],
           ('<', '\\'): ['^'],
           ('^', '\\'): ['<'],
           ('v', '\\'): ['>'],
           ('>', '/') : ['^'],
           ('<', '/') : ['v'],
           ('^', '/') : ['>'],
           ('v', '/') : ['<'],
           }

mirrors = DefaultDict(mirrors)

def v_add(x,y):
  return (x[0]+y[0],x[1]+y[1])

def energize(arr,coord):
  arr[coord[0]][coord[1]] = True
  return arr

def count(s):
  s0 = set()
  for x in s:
    s0.add(x[0])
  return len(s0)

def ray(start, d):
  q = [(start,d)]
  ex = set()
  while len(q) != 0:
    x = q.pop()
    ex.add(x)
    s = x[0]
    d = x[1]
    for d0 in mirrors[(d, map[s])]:
      v = dir[d0]
      x0 = v_add(x[0],v)
      if x0 in map.keys() and (x0,d0) not in ex:
        q.append((x0,d0))
  return count(ex)

ray((0,0),'>')
```

```python
def most_energy(s):
  c = defaultdict(lambda:0)
  m = 0
  for x in s:
    v = ray(x[0],x[1])
    if m < v:
      m = v
      pos = x
  return m, pos

def create_set():
  s = set()
  for i in range(N):
    s.add(((i,0),'>'))
    s.add(((i,M-1),'<'))
  for j in range(M):
    s.add(((0,j),'v'))
    s.add(((N-1,j),'^'))
  return s

s = create_set()

most_energy(s)
```
