```python
from collections import defaultdict
import heapq as hq
```

```python
f = open("test.txt", "r")

data = f.read().strip().split("\n")

f.close()
map = {}
N = len(data)
M = len(data[0])
for i in range(N):
  for j in range(M):
    map[(i,j)] = int(data[i][j])
```

```python
def v_add(x,y):
  return (x[0]+y[0],x[1]+y[1])

neighbours = {
    ''  : ['^', '>', 'v', '<'],
    '>' : ['>','^','v'],
    'v' : ['v', '>', '<'],
    '<' : ['<','^','v'],
    '^' : ['^', '>', '<']
}

dir = {
    '>' : (0,1),
    'v' : (1,0),
    '<' : (0,-1),
    '^' : (-1,0)
}

def neighbour(coord, d, count):
  nhd = []
  if count < 3:
    for d0 in neighbours[d]:
      v = dir[d0]
      coord0 = v_add(coord,v)
      if coord0 in map.keys():
        if d0 != d:
          nhd.append((coord0,d0,1))
        else:
          nhd.append((coord0,d0,count+1))
  else:
    for d0 in neighbours[d][1:]:
      v = dir[d0]
      coord0 = v_add(coord,v)
      if coord0 in map.keys():
        nhd.append((coord0,d0,1))
  return nhd

def taxi(x,y):
  return abs(x[0]-y[0]) + abs(x[1]-y[1])

def construct_path(pred, current):
  path = [current]
  while current in pred.keys():
    current = pred[current]
    path = [current] + path
  return path

def a_star(start, d, finish):
  pred = {}
  g_score = defaultdict(lambda:float('inf'))
  g_score[(start,d, 0)] = 0
  f_score = defaultdict(lambda:float('inf'))
  f_score[(start, d, 0)] = taxi(start,finish)
  open = []
  hq.heappush(open,(f_score[(start,d)],(start, d, 0)))
  while len(open) != 0:
    least, curr  = hq.heappop(open)
    if curr[0] == finish:
      return least, construct_path(pred,curr)
    for n in neighbour(*curr):
      g = g_score[curr] + map[n[0]]
      if g < g_score[n]:
        pred[n] = curr
        g_score[n] = g
        f_score[n] = g + taxi(n[0], finish)
        if n not in open:
          hq.heappush(open,(f_score[n], n))
  return 'fail'


a_star((0,0), '', (N-1,M-1))[0]

```
