```python
import re
```

```python
f = open("19-input.txt", "r")

data = f.read().strip().split("\n")

f.close()
```

```python
flows = {}
accepted  = 0
pattern = re.compile('<|>')
def interpret(part, flow):
  for inst in flow[:-1]:
    rule, send_to = inst.split(':')
    attr, val = pattern.split(rule)
    rule_type = rule.lstrip(attr).rstrip(val)
    val = int(val)
    part_val = part[attr]
    match rule_type:
      case '<':
        if part_val < val:
          return send_to
      case '>':
        if part_val > val:
          return send_to
  return flow[-1]

count  = 0

for l in data:
  if l == '':
    count += 1
    break
  i = l.index('{')
  flows[l[: i]] = l[i:].strip('{}').split(",")
  count += 1
rating  = 0
for l in data[count:]:
  part = {}
  l = l.strip("{}").split(",")
  for x in l:
    part[x[0]] = int(x[2:])
  send_to  = 'in'
  while send_to not in {'R','A'}:
    send_to = interpret(part,flows[send_to])
  if send_to == 'A':
    for k in part.keys():
      rating += part[k]



rating
```

```python
accept = []

p = {
    'x': [1,4001],
    'm': [1,4001],
    'a': [1,4001],
    's': [1,4001]
}

accept = [(p,'in')]

def interpret(part, flow):
  candidates = []
  for inst in flow[:-1]:
    rule, send_to = inst.split(':')
    attr, val = pattern.split(rule)
    rule_type = rule.lstrip(attr).rstrip(val)
    val = int(val)
    part_val = part[attr]
    match rule_type:
      case '<':
        if part_val[1] <= val:
          candidates.append(part,send_to)
          break
        elif part_val[0] < val:
          part0 = part.copy()
          part0[attr] = [part_val[0],val]
          part[attr] = [val, part_val[1]]
          candidates.append((part0,send_to))
      case '>':
        if part_val[0] >= val:
          candidates.append(part,send_to)
          break
        elif part_val[1] > val:
          part0 = part.copy()
          part0[attr] = [val + 1, part_val[1]]
          part[attr] = [part_val[0], val + 1]
          candidates.append((part0,send_to))
  candidates.append((part, flow[-1]))
  return candidates

total = 0
accepted = set()

while accept != []:
  p, f = accept.pop()
  prod = 1
  t = str(p.values())
  if f not in {'A','R'}:
    accept = accept + interpret(p,flows[f])
  elif f == 'A' and t not in accepted:
    prod = 1
    for k in p.keys():
      r = p[k]
      r = r[1] - r[0]
      prod *= r
    total += prod
    accepted.add(t)

total
```
