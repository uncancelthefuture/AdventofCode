```python
import re
import math
```

```python
f = open("/content/8-input", "r")

data = re.split("\n+",f.read())

f.close()

inst = data[0]

data = data[1:]
d = dict()

for l in data:
    k = re.search('(\w+)(?=\s\=)', l)
    if k:
      k = k.group(0)
    L = re.search('(?<=\()\w+(?=,)', l)
    if L:
      L = L.group(0)
    R = re.search('(?<=\s)\w+(?=\))', l)
    if R:
      R = R.group(0)
    d[(k,'L')] = L
    d[(k,'R')] = R

i = 0
N = len(inst)
pos = 'AAA'
while pos != 'ZZZ':
  j = i % N
  dir = inst[j]
  pos = d[(pos,dir)]
  i += 1
i

```

```python
pos = []
for k in d.keys():
  try:
    x = re.search("\w+A", k[0])
  except:
    TypeError
  if x and k[1] == 'L':
    pos.append(x.group())

print(pos)
vals = []
for m in range(len(pos)):
  i = 0
  while not re.search("\wZ", pos[m]):
      j = i % N
      dir = inst[j]
      pos[m] = d[(pos[m],dir)]
      i += 1
  vals.append((i))

vals
```

```python
from math import gcd
a = vals   #will work for an int array of any length
lcm = 1
for i in a:
    lcm = lcm*i//gcd(lcm, i)
print(lcm)
```
