```python
f  = open("18-input.txt", "r")

data = f.read().strip().split("\n")

f.close()

data = [x.split(" ") for x in data]

map = {}
```

```python
#solution is perimeter (p) + interior (i) points
#Pic's theorem: area = i + p/2
#area calc: SUM (y_{i+1}+y{i})*(x_{i+1}-x_{i})/2

p = 0
v = [(0,0)]

dir = {
    'R' : (1,0),
    'D' : (0,-1),
    'L' : (-1,0),
    'U' : (0,1)
}

def v_add(x,y):
  return (x[0] + y[0], x[1] + y[1])

last = v[-1]
for x in data:
  d, a = x[0:2]
  d = dir[d]
  p += int(a)
  for i in range(int(a)):
    last = v_add(last,d)
  v.append(last)

def area(v):
  N = len(v) - 1
  s = 0
  for i in range(N):
    x1 = v[i+1]
    x0 = v[i]
    s += (x1[1]+x0[1])*(x1[0]-x0[0])
  return s/2

def total(v,p):
  A = area(v)
  i = int(A - p/2 + 1)
  return i + p

total(v,p)
```
