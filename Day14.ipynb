```python
from collections import defaultdict

f = open("14-input.txt", "r")

data = [list(x) for x in f.read().strip().split("\n")]

f.close()

N = len(data[0])

dish = np.array(data)

def tilt(dish):
  h = len(dish)
  w = len(dish[0])
  for j in range(w):
    last = -1
    for i in range(h):
      x = dish[i][j]
      match x:
        case '#':
          last = i
        case 'O':
          if (last + 1)  != i:
            last += 1
            dish[last][j] = 'O'
            dish[i][j] = '.'
          else:
            last = i
  return dish


def weight(dish):
  dish0 = np.flip(dish,axis=0)
  h = len(dish)
  w = len(dish[0])
  s = 0
  for i in range(h):
    weight = (i+1)
    for j in range(w):
      if dish0[i][j] == 'O':
        s += weight
  return s

weight(tilt(dish))
```

```python
def rotate(dish):
  dish0 = np.rot90(dish,-1)
  return dish0

def cycle(dish):
  #suppose oriented to original
  dish0 = dish.copy()
  for i in range(4):
    dish0 = tilt(dish0)
    dish0 = rotate(dish0)
  return dish0

def hash(dish):
  return tuple(map(tuple, dish))

def run(dish, cycles):
  runs = {}
  dish0 = dish.copy()
  found = 0
  t = 0
  while not found:
    h = hash(dish0)
    if h not in runs.keys():
      runs[h] = t
    else:
      first_leg = runs[h]
      repeat_len = t - first_leg
      iter = (cycles - first_leg) % repeat_len
      cycles = t + iter
      found = 1
    dish0 = cycle(dish0)
    t += 1
  while t < cycles:
    dish0 = cycle(dish0)
    t += 1
  return dish0

weight(run(dish,1000000000))


```
