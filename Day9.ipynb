```python
import re

f = open("/content/9-input", "r")

data = f.read().strip()

f.close()

data = [list(map(int, x.strip().split(" "))) for x in data.split("\n")]
```

```python
def history(x):
  buffer = []
  for i in range(len(x) - 1):
    buffer.append(x[i + 1] - x[i])
  if buffer == [0]*(len(x)-1):
    return x[-1]
  else:
    return history(buffer) + x[-1]

s = 0

for x in data:
  s += history(x)
s
```

```python
def reverse_history(x):
  buffer = []
  for i in range(len(x) - 1):
    buffer.append(x[i + 1] - x[i])
  if buffer == [0]*(len(x)-1):
    return x[0]
  else:
    return x[0] - reverse_history(buffer)

s = 0

for x in data:
  s += reverse_history(x)
s
```
