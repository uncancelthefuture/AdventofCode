```python
import re
```

```python
f = open("15-input", "r")

data = f.read().strip()

f.close()

pattern = re.compile("\n|,")
data  = pattern.split(data)
```

```python
def hash(x):
  cv = 0
  for c in x:
    cv += ord(c)
    cv *= 17
    cv %= 256
  return cv

s = 0
for x in data:
  s += hash(x)
s

pattern = re.compile('(?=\=)|(?=\-)')

def hashmap(x):
  x = pattern.split(x)
  label = x[0]
  direction = x[1]
  return hash(label), label, direction

boxes = [[]]*(256)

def procedure(x, boxes):
  num, label, dir = hashmap(x)
  box = boxes[num]
  match dir[0]:
    case '=':
      lens = [label, int(dir[1])]
      replaced = False
      if box == []:
        box = [lens]
        replaced = True
      else:
        N = len(box)
        for i in range(N):
          if box[i][0] == label:
            box[i] = lens
            replaced = True
            break
        if not replaced:
          box.append(lens)
    case '-':
      N = len(box)
      for i in range(N):
        if box[i][0] == label:
          box.pop(i)
          break
  boxes[num] = box
  return boxes

def value(b):
  s = 0
  N = len(b)
  for i in range(N):
    weight = i + 1
    M = len(b[i])
    for j in range(M):
      s += weight * (j+1) * b[i][j][1]
  return s

for x in data:
  boxes = procedure(x, boxes)

value(boxes)
```
